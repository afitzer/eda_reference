```python
df['Hours'] = df['TIMES'].str.split('H').str[0].astype(int)
df['Minutes'] = df['TIMES'].str.split('H').str[1].str.split('\'').str[0].astype(int)
df['Seconds'] = df['TIMES'].str.split('H').str[1].str.split('\'').str[1].str.split('\"').str[0].astype(int)
```

```python
# concat the hours, minutes, seconds and then convert to timedelta
df['Time Join'] = pd.to_timedelta(df['Hours'], unit='h') + pd.to_timedelta(df['Minutes'], unit='m') + pd.to_timedelta(df['Seconds'], unit='s')
df
```

```python
# Custom function to format Timedelta
def format_timedelta(timedelta):
    total_seconds = timedelta.total_seconds()
    hours = int(total_seconds // 3600)
    minutes = int((total_seconds % 3600) // 60)
    seconds = int(total_seconds % 60)
    return f"{hours:02d}:{minutes:02d}:{seconds:02d}"

# Apply the custom function to format the column
df['Time Join'] = df['Time Join'].apply(format_timedelta)
```

```python
# Custom function to convert formatted string to Timedelta
def string_to_timedelta(formatted_str):
    hours, minutes, seconds = map(int, formatted_str.split(':'))
    return pd.Timedelta(hours=hours, minutes=minutes, seconds=seconds)

# Apply the custom function to convert the column
df['Time Join'] = df['Time Join'].apply(string_to_timedelta)
```
