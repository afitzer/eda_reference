```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read in the passing data
df = pd.read_csv('data/team_stats10-16-23.csv')
```

```python
df.info()
```

```python
df
```

```python
fig, ax = plt.subplots(1,2,figsize=(15,5))
sns.histplot(df['Passing Completion %'],ax=ax[0])
sns.histplot(df['Passing Yards/Attempt'],ax=ax[1])
```

```python
sns.relplot(data=df, x='Pass Yds', y='Passing Completions')
plt.title('Passing Yds vs Completions')
```

```python
# select passing completion % and team from the df and sort by passing completion %%!
PC_Perc = df[['Team', 'Passing Completion %', 'Passing Rate']].sort_values(by='Passing Completion %', ascending=False)

fig, ax = plt.subplots(figsize=(15, 5))
sns.barplot(data=PC_Perc, x='Team', y='Passing Completion %', palette='dark')
plt.xticks(rotation=90)
plt.title('Passing Completion % by Team')
plt.show()
```
