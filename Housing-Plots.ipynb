```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "Built 2020 or later": 920272287900,
    "Built 2010 to 2019": 429725783840,
    "Built 2000 to 2009": 613099675880,
    "Built 1990 to 1999": 479479696030,
    "Built 1980 to 1989": 419888274150,
    "Built 1970 to 1979": 417132593330,
    "Built 1960 to 1969": 327288145820,
    "Built 1950 to 1959": 340789273030,
    "Built 1940 to 1949": 139735783560,
    "Built 1939 or earlier": 4012747988}
```

```python
s = pd.Series(data, name="Housing Units")
```

```python
# plot a bar chart with the series s
# make the plot size 15, 5)
# set the title to "Housing Units by Year Built"
# set the x-axis label to "Year Built"
# set the y-axis label to "Housing Units"
fig, ax = plt.subplots(figsize=(15, 5))
sns.barplot(x=s.index, y=s)
plt.title("Housing Units by Year Built")
plt.xlabel("Year Built")
plt.ylabel("Housing Units")
# angle the x-axis labels 45 degrees
plt.xticks(rotation=45)
plt.show()
```
